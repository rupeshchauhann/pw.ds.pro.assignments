```python
import random
import math
from collections import Counter

# Generate a list of 100 integers containing values between 90 to 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Calculate the mean of a given list of numbers
def calc_mean(numbers):
    return sum(numbers) / len(numbers)

# (ii) Compute the mode of a list of integers
def calc_mode(numbers):
    frequency = Counter(numbers)
    max_count = max(frequency.values())
    modes = [key for key, val in frequency.items() if val == max_count]
    return modes

# (iii) Calculate the weighted mean of a list of values and their corresponding weights
def calc_weighted_mean(values, weights):
    if len(values) != len(weights):
        raise ValueError("Values and weights must have the same length.")
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weights = sum(weights)
    return weighted_sum / total_weights

# (iv) Find the geometric mean of a list of positive numbers
def calc_geometric_mean(numbers):
    if any(x <= 0 for x in numbers):
        raise ValueError("All numbers must be positive.")
    product = math.prod(numbers)
    return product ** (1 / len(numbers))

# (v) Calculate the harmonic mean of a list of values
def calc_harmonic_mean(numbers):
    if any(x <= 0 for x in numbers):
        raise ValueError("All numbers must be positive.")
    reciprocal_sum = sum(1 / x for x in numbers)
    return len(numbers) / reciprocal_sum

# (vi) Determine the midrange of a list of numbers
def calc_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Find the trimmed mean of a list, excluding a certain percentage of outliers
def calc_trimmed_mean(numbers, trim_percent):
    if not (0 <= trim_percent < 50):
        raise ValueError("Trim percent must be between 0 and 50.")
    numbers_sorted = sorted(numbers)
    trim_count = int(len(numbers) * trim_percent / 100)
    trimmed_numbers = numbers_sorted[trim_count:len(numbers) - trim_count]
    return sum(trimmed_numbers) / len(trimmed_numbers)

# Example usage
print("List:", int_list)
print("Mean:", calc_mean(int_list))
print("Median:", sorted(int_list)[len(int_list) // 2])
print("Mode:", calc_mode(int_list))

weights = [random.randint(1, 10) for _ in range(100)]
print("Weighted Mean:", calc_weighted_mean(int_list, weights))
print("Geometric Mean:", calc_geometric_mean([x for x in int_list if x > 0]))
print("Harmonic Mean:", calc_harmonic_mean([x for x in int_list if x > 0]))
print("Midrange:", calc_midrange(int_list))
print("Trimmed Mean (10%):", calc_trimmed_mean(int_list, 10))
```

```python
import random
import math
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

# Generate a list of 100 integers containing values between 90 to 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Generate a list of 500 integers containing values between 200 to 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# (i) Compare visualizations for the given data
def compare_visualizations(data):
    plt.figure(figsize=(15, 5))

    # Frequency & Gaussian distribution
    plt.subplot(1, 3, 1)
    sns.histplot(data, kde=False, stat="density", bins=30, color="blue", label="Frequency")
    sns.kdeplot(data, color="red", label="Gaussian")
    plt.title("Frequency & Gaussian Distribution")
    plt.legend()

    # Frequency smoothened KDE plot
    plt.subplot(1, 3, 2)
    sns.histplot(data, kde=True, stat="density", bins=30, color="green")
    plt.title("Frequency Smoothened KDE Plot")

    # Gaussian distribution & smoothened KDE plot
    plt.subplot(1, 3, 3)
    sns.kdeplot(data, color="purple", label="Gaussian")
    sns.kdeplot(data, bw_adjust=0.5, color="orange", label="Smoothened KDE")
    plt.title("Gaussian Distribution & Smoothened KDE Plot")
    plt.legend()

    plt.tight_layout()
    plt.show()

# (ii) Calculate the range of a given list of numbers
def calc_range(numbers):
    return max(numbers) - min(numbers)

# (iii) Find the variance and standard deviation of a list of numbers
def calc_variance(numbers):
    mean = calc_mean(numbers)
    return sum((x - mean) ** 2 for x in numbers) / len(numbers)

def calc_std_dev(numbers):
    return math.sqrt(calc_variance(numbers))

# (iv) Compute the interquartile range (IQR) of a list of values
def calc_iqr(numbers):
    numbers_sorted = sorted(numbers)
    q1 = numbers_sorted[len(numbers_sorted) // 4]
    q3 = numbers_sorted[(3 * len(numbers_sorted)) // 4]
    return q3 - q1

# (v) Calculate the coefficient of variation for a dataset
def calc_coefficient_of_variation(numbers):
    mean = calc_mean(numbers)
    std_dev = calc_std_dev(numbers)
    return (std_dev / mean) * 100

# (vi) Find the mean absolute deviation (MAD) of a list of numbers
def calc_mad(numbers):
    mean = calc_mean(numbers)
    return sum(abs(x - mean) for x in numbers) / len(numbers)

# (vii) Calculate the quartile deviation of a list of values
def calc_quartile_deviation(numbers):
    return calc_iqr(numbers) / 2

# (viii) Find the range-based coefficient of dispersion for a dataset
def calc_range_coefficient_of_dispersion(numbers):
    return (max(numbers) - min(numbers)) / (max(numbers) + min(numbers))

# Example usage
print("List:", int_list)
print("Mean:", calc_mean(int_list))
print("Median:", sorted(int_list)[len(int_list) // 2])
print("Mode:", calc_mode(int_list))

weights = [random.randint(1, 10) for _ in range(100)]
print("Weighted Mean:", calc_weighted_mean(int_list, weights))
print("Geometric Mean:", calc_geometric_mean([x for x in int_list if x > 0]))
print("Harmonic Mean:", calc_harmonic_mean([x for x in int_list if x > 0]))
print("Midrange:", calc_midrange(int_list))
print("Trimmed Mean (10%):", calc_trimmed_mean(int_list, 10))

# Additional calculations for int_list2
print("Range:", calc_range(int_list2))
print("Variance:", calc_variance(int_list2))
print("Standard Deviation:", calc_std_dev(int_list2))
print("IQR:", calc_iqr(int_list2))
print("Coefficient of Variation:", calc_coefficient_of_variation(int_list2))
print("MAD:", calc_mad(int_list2))
print("Quartile Deviation:", calc_quartile_deviation(int_list2))
print("Range Coefficient of Dispersion:", calc_range_coefficient_of_dispersion(int_list2))

# Visualization for int_list2
compare_visualizations(int_list2)
```

```python
import random
import numpy as np

# 3) Class representing a discrete random variable
class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        """
        Initialize the random variable.
        :param outcomes: List of possible outcomes
        :param probabilities: Corresponding probabilities for each outcome
        """
        if len(outcomes) != len(probabilities):
            raise ValueError("Outcomes and probabilities must have the same length.")
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("Probabilities must sum to 1.")
        
        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        """Calculate the expected value."""
        return sum(o * p for o, p in zip(self.outcomes, self.probabilities))

    def variance(self):
        """Calculate the variance."""
        mean = self.expected_value()
        return sum(p * (o - mean) ** 2 for o, p in zip(self.outcomes, self.probabilities))


# 4) Simulate rolling a six-sided die and calculate expected value and variance

def simulate_die_rolls(n_rolls=1000):
    """Simulate rolling a fair six-sided die."""
    die_outcomes = [1, 2, 3, 4, 5, 6]
    die_probabilities = [1/6] * 6

    # Simulate rolls
    rolls = [random.choice(die_outcomes) for _ in range(n_rolls)]

    # Calculate expected value and variance from simulation
    simulated_mean = np.mean(rolls)
    simulated_variance = np.var(rolls, ddof=0)

    # Calculate theoretical expected value and variance
    die = DiscreteRandomVariable(die_outcomes, die_probabilities)
    theoretical_mean = die.expected_value()
    theoretical_variance = die.variance()

    return simulated_mean, simulated_variance, theoretical_mean, theoretical_variance

# 5) Generate random samples and calculate mean and variance

def generate_samples_and_stats(distribution, params, n_samples=1000):
    """Generate samples from a given distribution and calculate stats."""
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, n_samples)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, n_samples)
    else:
        raise ValueError("Unsupported distribution")

    mean = np.mean(samples)
    variance = np.var(samples, ddof=0)
    return samples, mean, variance

# 6) Generate random numbers from Gaussian distribution

def gaussian_stats(mean=0, std_dev=1, n_samples=1000):
    """Generate random numbers from Gaussian distribution and compute stats."""
    samples = np.random.normal(mean, std_dev, n_samples)
    computed_mean = np.mean(samples)
    computed_variance = np.var(samples, ddof=0)
    computed_std_dev = np.std(samples, ddof=0)
    return samples, computed_mean, computed_variance, computed_std_dev

# Example usage
if __name__ == "__main__":
    # Part 4: Simulate die rolls
    sim_mean, sim_var, theo_mean, theo_var = simulate_die_rolls()
    print(f"Simulated Mean: {sim_mean}, Simulated Variance: {sim_var}")
    print(f"Theoretical Mean: {theo_mean}, Theoretical Variance: {theo_var}")

    # Part 5: Generate samples for binomial distribution
    samples, mean, variance = generate_samples_and_stats('binomial', (10, 0.5))
    print(f"Binomial Mean: {mean}, Variance: {variance}")

    # Part 6: Gaussian stats
    samples, mean, variance, std_dev = gaussian_stats(0, 1)
    print(f"Gaussian Mean: {mean}, Variance: {variance}, Std Dev: {std_dev}")
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset('tips')

# Function to calculate skewness
def calculate_skewness(column):
    return column.skew()

# Function to determine skewness type
def skewness_type(column):
    skewness = calculate_skewness(column)
    if skewness > 0:
        return "Positive skewness"
    elif skewness < 0:
        return "Negative skewness"
    else:
        return "Approximately symmetric"

# Function to calculate covariance
def calculate_covariance(column1, column2):
    return np.cov(column1, column2, bias=True)[0][1]

# Function to calculate Pearson correlation coefficient
def calculate_pearson_correlation(column1, column2):
    return np.corrcoef(column1, column2)[0][1]

# Visualization of correlation using scatter plot
def plot_correlation(df, column1, column2):
    plt.figure(figsize=(8, 6))
    plt.scatter(df[column1], df[column2], alpha=0.6, edgecolors='k')
    plt.title(f'Scatter Plot: {column1} vs {column2}', fontsize=14)
    plt.xlabel(column1, fontsize=12)
    plt.ylabel(column2, fontsize=12)
    plt.grid(True)
    plt.show()

# Main program execution
if __name__ == "__main__":
    total_bill = data['total_bill']
    tip = data['tip']

    # (i) Calculate skewness
    total_bill_skewness = calculate_skewness(total_bill)
    tip_skewness = calculate_skewness(tip)
    print(f"Skewness of total_bill: {total_bill_skewness}")
    print(f"Skewness of tip: {tip_skewness}")

    # (ii) Determine skewness type
    print(f"Skewness type of total_bill: {skewness_type(total_bill)}")
    print(f"Skewness type of tip: {skewness_type(tip)}")

    # (iii) Calculate covariance
    covariance = calculate_covariance(total_bill, tip)
    print(f"Covariance between total_bill and tip: {covariance}")

    # (iv) Calculate Pearson correlation coefficient
    pearson_correlation = calculate_pearson_correlation(total_bill, tip)
    print(f"Pearson correlation coefficient between total_bill and tip: {pearson_correlation}")

    # (v) Visualize correlation
    plot_correlation(data, 'total_bill', 'tip')
```

#### 8. PDF for a Normal Distribution

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.
    
    Parameters:
    - x: Value at which to evaluate the PDF.
    - mean: Mean () of the distribution.
    - std_dev: Standard deviation () of the distribution.
    
    Returns:
    - PDF value at x.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    return coefficient * math.exp(exponent)

# Example usage:
print(normal_pdf(1, 0, 1))  # Standard normal distribution
```

#### 9. CDF for an Exponential Distribution

```python
import math

def exponential_cdf(x, rate):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution.
    
    Parameters:
    - x: Value at which to evaluate the CDF (x >= 0).
    - rate: Rate parameter () of the distribution ( > 0).
    
    Returns:
    - CDF value at x.
    """
    if x < 0:
        return 0  # Exponential CDF is 0 for x < 0
    return 1 - math.exp(-rate * x)

# Example usage:
print(exponential_cdf(2, 1))  # Exponential distribution with =1
```

#### 10. PMF for a Poisson Distribution

```python
import math

def poisson_pmf(k, rate):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution.
    
    Parameters:
    - k: Number of events (k >= 0, integer).
    - rate: Average rate of occurrence ().
    
    Returns:
    - PMF value for k.
    """
    if k < 0 or not isinstance(k, int):
        return 0  # PMF is 0 for negative or non-integer k
    return (rate ** k * math.exp(-rate)) / math.factorial(k)

# Example usage:
print(poisson_pmf(3, 2))  # Poisson distribution with =2
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate the number of successes and total observations
successes = [old_layout.sum(), new_layout.sum()]
n_obs = [len(old_layout), len(new_layout)]

# Perform a z-test
z_stat, p_value = proportions_ztest(successes, n_obs, alternative='larger')

# Output results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:  # Assuming a significance level of 0.05
    print("Reject the null hypothesis: The new layout leads to a higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the new layout is better.")
```

```python
import numpy as np
from scipy.stats import norm

# Generate the data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate differences
differences = after_program - before_program

# Calculate statistics
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the z-statistic
z_stat = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value (one-tailed test)
p_value = 1 - norm.cdf(z_stat)

# Output results
print(f"Mean Difference: {mean_diff:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:  # Assuming a significance level of 0.05
    print("Reject the null hypothesis: The tutoring program significantly improves scores.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the tutoring program improves scores.")
```

```python
import numpy as np
from scipy.stats import t

def ab_test_analysis(layout_a, layout_b, alpha=0.05):
    # Calculate sample statistics
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    std_a = np.std(layout_a, ddof=1)  # Sample standard deviation
    std_b = np.std(layout_b, ddof=1)
    n_a = len(layout_a)
    n_b = len(layout_b)
    
    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((n_a - 1) * std_a**2 + (n_b - 1) * std_b**2) / (n_a + n_b - 2))
    
    # Calculate the t-statistic
    t_stat = (mean_a - mean_b) / (pooled_std * np.sqrt(1/n_a + 1/n_b))
    
    # Calculate the degrees of freedom
    df = n_a + n_b - 2
    
    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - t.cdf(np.abs(t_stat), df))
    
    # Output results
    print(f"Mean of Layout A: {mean_a:.2f}")
    print(f"Mean of Layout B: {mean_b:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")
    
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the layouts.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the layouts.")

# Data for the layouts
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Run the A/B test analysis
ab_test_analysis(layout_a_clicks, layout_b_clicks)
```

```python
import numpy as np
from scipy.stats import t

def intervention_analysis(pre_scores, post_scores, alpha=0.05):
    # Calculate the differences between post and pre-intervention scores
    differences = np.array(post_scores) - np.array(pre_scores)
    
    # Calculate sample statistics for the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)  # Sample standard deviation
    n = len(differences)  # Number of pairs
    
    # Calculate the t-statistic
    t_stat = mean_diff / (std_diff / np.sqrt(n))
    
    # Calculate the degrees of freedom
    df = n - 1
    
    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - t.cdf(np.abs(t_stat), df))
    
    # Output results
    print(f"Mean of Differences: {mean_diff:.2f}")
    print(f"Standard Deviation of Differences: {std_diff:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")
    
    if p_value < alpha:
        print("Reject the null hypothesis: The intervention had a significant impact.")
    else:
        print("Fail to reject the null hypothesis: No significant impact from the intervention.")

# Data for pre- and post-intervention scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Run the intervention analysis
intervention_analysis(pre_intervention_scores, post_intervention_scores)
```

```python
import numpy as np
from scipy.stats import t

def salary_gap_analysis(male_salaries, female_salaries, alpha=0.05):
    # Calculate sample statistics for male salaries
    mean_male = np.mean(male_salaries)
    std_male = np.std(male_salaries, ddof=1)
    n_male = len(male_salaries)
    
    # Calculate sample statistics for female salaries
    mean_female = np.mean(female_salaries)
    std_female = np.std(female_salaries, ddof=1)
    n_female = len(female_salaries)
    
    # Calculate the t-statistic
    pooled_std = np.sqrt(((n_male - 1) * std_male**2 + (n_female - 1) * std_female**2) / (n_male + n_female - 2))
    t_stat = (mean_male - mean_female) / (pooled_std * np.sqrt(1/n_male + 1/n_female))
    
    # Calculate the degrees of freedom
    df = n_male + n_female - 2
    
    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - t.cdf(np.abs(t_stat), df))
    
    # Output results
    print(f"Mean Male Salary: ${mean_male:.2f}")
    print(f"Mean Female Salary: ${mean_female:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")
    
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference in average salaries between male and female employees.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in average salaries between male and female employees.")

# Generate synthetic salary data
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Run the salary gap analysis
salary_gap_analysis(male_salaries, female_salaries)
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a contingency table
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Output results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("\nFail to reject the null hypothesis: No significant association between age groups and voter preferences.")
```
